```python
#1 Tokenisation
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
text = "I have AMT component today"
tokens = word_tokenize(text)
print(tokens)
```

```python
# Sentense Tokenization
from nltk.tokenize import sent_tokenize
text = "Complex Nos is Hard, It is widely used in Mathematics"
sentences = sent_tokenize(text)
print(sentences)
```

```python
# Stop word removal
# Stop Words as is, the, and, in are removed.

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
filtered_words = [w for w in words if w.lower() not in stop_words]
print(filtered_words)
```

```python
# Stemming
# Stemming Reduces words to their root form
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["playing", "studies", "running", "gone", "watched"]
for w in words:
  print(stemmer .stem(w))
```

```python
# lemmatization
# Lemmatization converts words to their dictionary form
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("writing"))
print(lemmatizer.lemmatize("cacti"))
print(lemmatizer.lemmatize("geese"))
```

```python
# Part of Speech (pos) Tagging
# POS tagging identifies grammatical roles of words
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize # Added this import as word_tokenize is used but not imported in this specific cell.
words = word_tokenize("Python is a powerful programming language")
pos_tags = pos_tag(words)
print(pos_tags)
```
